### Решение модельного уравнения

Необходимо решить модельное уравнение:
$$
\begin{align}
\label{eq:dalq}
u'(t) = \lambda u(t)
\end{align}
$$
с начальным условием $u(0) = 1$ с помошью явного и неявного метода Эйлера. Считать, что $\lambda = -200$. Провести расчеты на сетке с числом узлов $N = 1001; 101; 100$. Построить графики решения на одном рисунке, объяснить результат. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ex_euler(u, l, h, N):
    for i in range(0, N-1):
        u[i+1] = (1+l*h)*u[i]

def im_euler(u, l, h, N):
    for i in range(0, N-1):
 # TODO : реализуйте неявный метод Эйлера для модельного уравнен
        u[i+1] = u[i]/(1-l*h)
    
T = 1 # считае до этого момента
N = 100

x_teor = np.linspace(0,T,(N-1)*100+1) # сетка для точного решения
l = -200

def gridplot(l, T, N, method):
    h = T/(N-1)
    u = np.zeros(N)
    u[0] = 1 #начальное условие
    x = np.linspace(0,T,N) # сетка

    if method == 'ex':
        ex_euler(u, l, h, N)
    elif method == 'im':
        im_euler(u, l, h, N)

    plt.plot(x, u, 'o', label=str(N))



plt.title('explicit')
plt.plot(x_teor, np.exp(l*x_teor), linewidth=5.0, label='exact')
gridplot(l,T,100, 'ex')
gridplot(l,T,101, 'ex')
gridplot(l,T,1001, 'ex')

plt.legend()
plt.axis([0, 0.12, -1.3, 1.3])
```

```python
plt.title('implicit')
plt.plot(x_teor, np.exp(l*x_teor), linewidth=5.0, label='exact')
gridplot(l,T,100, 'im')
gridplot(l,T,101, 'im')
gridplot(l,T,1001, 'im')

plt.legend()
plt.axis([0, 0.12, -0.2, 1.2])
```
